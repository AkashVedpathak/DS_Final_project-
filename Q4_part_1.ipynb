```python
import numpy as np
import pandas as pd
```

```python

# Load and clean data
df = pd.read_csv(r"/content/Batchwise Attendance Data - Class 1.csv")
```

```python
df.isna().sum()
```

```python
df['Type']=df['Type'].fillna('STUDENT')
```

```python
df=df.fillna("Missed")
```

```python
stud_count = df[df.Type == 'STUDENT'].shape[0]
stud_count
```

```python
stud_df = df[df["Type"].str.contains("STUDENT") == True]
stud_df
```

```python
stud_df.shape
```

```python
Attended_count = (stud_df == 'Attended').astype(object).sum(axis=1)
```

```python
Attended_count.info()
```

```python
frame = {"Attended Count": Attended_count}
```

```python
a_df = pd.DataFrame(frame)
```

```python
a_df = a_df['Attended Count'] = a_df['Attended Count'].astype(int)
```

```python
a_df.head()
```

```python
stud_df['Attended Count'] = a_df
```

```python
stud_df.head(5)
```

```python
b=stud_df.select_dtypes(exclude='object')
```

```python
b
```

```python
k=b.drop(['Student Roll Num'],axis=1)
```

```python
k
```

```python
rating_count = k.sum(axis = 1)
```

```python
rating_count_avg = rating_count/100*100
```

```python
rating_count_avg.astype(int)
```

```python
stud_df['rating_count_avg']= rating_count_avg
```

```python
stud_df.head()
```

```python
stud_dff = stud_df['rating_count_avg'] = stud_df['rating_count_avg'].astype(int)
```

```python
stud_df
```

```python
import matplotlib.pyplot as plt
```

```python
low_attendance = stud_df[stud_df['Attended Count'] <= 4]
low_attendance
```

```python

# Assuming that the column representing attendance is named "attendance"
low_attendance = stud_df[stud_df['Attended Count'] <= 4]


low_attendance.plot(kind='bar', x='Student Roll Num', y='Attended Count', color='red')

plt.title("Students with low attendance")
plt.xlabel("Student Roll Number")
plt.ylabel("Attendance less than 4")

plt.show()



#more attension
```

```python
low_attendance = stud_df[stud_df['rating_count_avg'] <= 70]
low_attendance.plot(kind='bar', x='Student Roll Num', y='rating_count_avg', color='red')
plt.title("Students are Unsatisfied")
plt.xlabel("Student Roll Number")
plt.ylabel("Rating count avg")
plt.show()
```

```python

low_attendance = stud_df[stud_df['rating_count_avg'] > 80]
low_attendance.plot(kind='bar', x='Student Roll Num', y='rating_count_avg', color='red')
plt.title("Students are Very Satisfied")
plt.xlabel("Student Roll Number")
plt.ylabel("Rating count avg")

plt.show()
```

```python
# averge Student ratng
low_attendance = stud_df[(stud_df['rating_count_avg'] >= 75) & (stud_df['rating_count_avg']<=80)]
low_attendance.plot(kind='bar', x='Student Roll Num', y='rating_count_avg', color='blue')
plt.title("Students are average performace ")
plt.xlabel("Student Roll Number")
plt.ylabel("Rating count avg")

#axs[1]plt.show()
```
